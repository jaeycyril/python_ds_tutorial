## DATA STRUCTURES

### 1. LISTS

```python
# LIST CREATION

first_list = [1, 2, 'Sunflower', '2.5']
second_list = ["Hello", ',', 'World']
third_list = list(range(1,10))

print(first_list)
print(second_list)
print(third_list)
```

```python
# LIST INDEXING

first_element = second_list[0]   # First element in the list
third_element = second_list[2]   # Third element of the list

print(first_element)
print(third_element)
```

```python
# LIST SLICING

slice1 = third_list[3:10:2]  # Start at item 3. Stop at item 10. Skip 2 items at a time 
slice2 = third_list[3:7]     # Start at item 3. Stop at item 7. Do not skip any item

print(slice1)
print(slice2)
```

```python
# LIST APPEND

first_list.append("Rose")

first_list
```

## FUNCTIONS

```python
# FUNCTION DEFINITION

def sum_up(a_list):
    """
    A function to sum up all the numbers in a list
    param: List of numbers
    returns: sumof numbers in list
    """
    sum_all = 0
    for i in a_list:
        sum_all += i
    return sum_all

# FUNCTION CALL

sample_list = [20, 40, 60, 80, 100]
sum_up(sample_list)
```

```python
# SOME FUNCTION TAKE ACTIONS BUT DO NOT RETURN ANY VALUES

def print_pretty(word):
    print('*'*50)
    print(word)
    print('='*50)

print_pretty('Troy and Abed in the morning')
```

```python
# LAMBDA FUNCTIONS
# These are anonymous functions which perform one expression and return the result

change_to_lower_case = lambda x: x.lower() # Define the lambda function

change_to_lower_case("THIS SHOULD BE IN LOWER CASE") # Call it
```

## DATA MANIPULATION WITH PANDAS

Pandas is a wonderful library developed by Wes Mckinney for Data Manipulation.<br>

We will go through the following Data Manipulation Verbs to provide an understanding<br>
of the basic capabilities of Pandas.
<ul>
    <li>Select</li>
    <li>Filter</li>
    <li>Aggregate</li>
    <li>Sort</li>
    <li>Mutate</li>
    <li>Combine</li>
</ul>

Look at the Pandas [documentation](https://pandas.pydata.org/docs/getting_started/) for more details.

```python
import pandas as pd
```

```python
# First up, let us read in some data. 
# Here, we read a CSV file but Pandas is capable of handling many other formats

dataset = pd.read_csv("adult_v1.csv")
dataset.head()
```

### 1. SELECT
This is basically getting or or more columns from the dataset

```python
# Use loc for Selecting columns by their names
# Select Age column
# Pandas does not show all the data as a default behaviour

dataset.loc[:, ["age"]]
```

```python
# Select Age, Sex and Native Country
dataset.loc[:, ["age", "sex", "native-country"]]
```

```python
# Use iloc for positional column values
# Example: SELECT age (which is in position 0)

dataset.iloc[:, 0]
```

```python
# Use dictionary notation
dataset['age']
```

### 2. FILTER
Limit the number of rows returned

```python
# Limit the number of rows using indexing

# Return the 30th to the 35th rows

dataset.loc[30:35, :]
```

```python
# Use indexing and select

dataset.loc[30:35, ["age", "race"]]
```

```python
# Use Boolean masks to filter

# Query: Get all adults aged 70 and above

elderly_subset = dataset[(dataset['age']>=70)]
print(elderly_subset.shape) # (Number of rows, Number of columns)
elderly_subset.head()
```

```python
# Masks can be as complex as you want
# Query: Get all Female adults from the United States with Masters or Some College education

college_subset = dataset[(dataset['sex'] == "Female") & (dataset['native-country'] == "United-States")
                        & ((dataset['education'] == "Masters") | (dataset["education"] == "Some-college"))]

college_subset
```

### 3. AGGREGATE
This simply refers to summarising columns or rows or data into a single value e.g. Average or Count

```python
# Summary stats
dataset.describe()
```

```python
# Column aggregation
# Average Age

dataset['age'].mean()
```

```python
dataset.mean() # Gets the numerical columns and does the calculations
```

```python
# We can also Group by values and aggregate

# What is average age of Male and Female adults in the dataset?

dataset.groupby('sex').aggregate({'age':'mean'})
```

```python
# Can also group by more than one column and aggregator

dataset.groupby(['sex','race']).aggregate({'age':'mean', 'education-num':'count'})
```

### 4. MUTATE
This is a fancy word for the process of changing the dataset e.g by adding a new column

```python
# Derived columns
# Gain_hrs column is a product of capital-gain column and horus-per-week column

dataset['Gain_hrs'] = dataset['capital-gain'] * dataset['hours-per-week']
dataset.head()
```

```python
# Use a function to derive a new column from an old column

def encode_sex(sex):
    if sex == "Female":
        return 1
    return 0

dataset['sex_code'] = dataset['sex'].apply(encode_sex)

dataset.head()
```

```python
# Delete columns
# axis=1 for column. Omit that to drop rows insteadd
dataset = dataset.drop(['Gain_hrs', 'sex_code'], axis=1)
dataset.head()
```

### SORT

```python
# Sort by 1 or more columns
# Age is sorted in Descending order
# Hours-per-week in Ascending order

dataset.sort_values(by=["age","hours-per-week"], ascending=[False, True]).head()
```

### COMBINE
This is simply combining datasets by concatenating or joining them

```python
# Concatenating 

df_sample1 = dataset.sample(20) # This is a sample of 20 rows from our dataset
df_sample2 = dataset.sample(30) # sample of 30 rows

df_samples_concatenated = pd.concat([df_sample1, df_sample2]) # This appends sample 2 to the bottom of sample 1
print(df_samples_concatenated.shape)
df_samples_concatenated.head()
```

```python
# For JOINING we create another dataset
occ_levels = {"occupation":['Adm-clerical', 'Exec-managerial', 'Handlers-cleaners',
               'Prof-specialty', 'Other-service', 'Sales', 'Craft-repair',
               'Transport-moving', 'Farming-fishing', 'Machine-op-inspct',
               'Tech-support', 'Protective-serv', 'Armed-Forces',
               'Priv-house-serv'],
                "occupation_level":['level 2', 'level 13', 'level 1',
                                    'level 12', 'unknown', 'level 7', 'level 4',
                                    'level 5', 'level 3', 'level 6',
                                    'level 11', 'level 8', 'level 9',
                                    'level 10']}

occ_level_df = pd.DataFrame(occ_levels)
occ_level_df
```

```python
# We can JOIN the datasets using the common columns sometimes know as foreign keys
# An INNER JOIN only returns rows that have matches in both datasets.
# Notice the reduction in number of rows in the result of the INNER JOIN

merged_data = pd.merge(dataset, occ_level_df, on="occupation", how="inner")
print(merged_data.shape)
merged_data.head()
```
